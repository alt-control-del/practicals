Write a program to construct a Bayesian network considering medical data. Use this model to demonstrate
the diagnosis of heart patients using the standard Heart Disease Data Set

```python
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import ParameterEstimator, BayesianEstimator
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
data = pd.read_csv(r'C:\Users\Tejas\OneDrive\Desktop\College\Practicals\CL 2\IR\Dataset\heart_cleveland_upload.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
model = BayesianNetwork([('age', 'condition'),
('sex', 'condition'),
('cp', 'condition'),
('trestbps', 'condition'),
('chol', 'condition')])
```

```python
model.fit(data,estimator=MaximumLikelihoodEstimator)
```

```python
HeartDiseasetest_infer = VariableElimination(model)
```

```python
q1=HeartDiseasetest_infer.query(variables=['condition'],evidence={'cp':3})
print(q1)
```

```python
q2=HeartDiseasetest_infer.query(variables=['condition'],evidence={'trestbps':140})
print(q2)
```

```python
q3=HeartDiseasetest_infer.query(variables=['condition'],evidence={'age':50})
print(q3)
```

```python
q4=HeartDiseasetest_infer.query(variables=['condition'],evidence={'chol':220})
print(q4)
```
